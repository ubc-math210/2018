## 1. Packages and modules

A Python [module](https://docs.python.org/3/tutorial/modules.html) is a text file containing Python code (with extension .py). A Python [package](https://docs.python.org/3/tutorial/modules.html#packages) is simply a collection of modules. Think of modules as files and packages as folders. We use the import keyword to import a package and use the dot syntax to access the modules, functions and variables inside a package.

## 3. Matplotlib

Let's plot these arrays using [matplotlib](https://matplotlib.org/)! Import the [pyplot module](https://matplotlib.org/api/pyplot_summary.html) and use [Jupyter magic](http://ipython.readthedocs.io/en/stable/interactive/magics.html) to display the figures in the notebook:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's plot $y = e^{-0.1x^2} \cos(2 \pi x)$ for $x \in [-10,10]$.

```python
x = np.linspace(-10,10,500)
y = np.exp(-0.1*x**2) * np.cos(2*np.pi*x)
plt.plot(x,y)
plt.show()
```

pyplot commands add features to the current figure. The command `plt.show()` renders the current figure. The next pyplot command will then create a new figure!

```python
x = np.linspace(-10,10,200)
y = np.exp(-0.1*x**2) * np.cos(2*np.pi*x)
plt.plot(x,y,'g-.')
plt.show()

x = np.linspace(-10,10,200)
y = np.exp(-0.1*x**2) * np.cos(2*np.pi*x)
plt.plot(x,y,'r-.')
plt.show()
```

## 4. SciPy

[SciPy](https://scipy.org/) is a collection of modules and subpackages for scientific computing including ODE solvers, optimization, statistics, etc. Let's import the subpackage `special` and plot the error function.

```python
from scipy import special
```

```python
special.erf?
```

```python
x = np.linspace(-5,5,100)
y = special.erf(x)
plt.plot(x,y)
plt.show()
```

## 5. pandas

[pandas](http://pandas.pydata.org/) is a Python package for data analysis. The core pandas object is a `DataFrame`. It's like a spreadsheet!

```python
import pandas as pd
```

Let's import a `.csv` file with crime statistics from the [Vancouver Police Department](http://data.vancouver.ca/datacatalogue/crime-data.htm). The file is saved in a folder on my website.

```python
data = pd.read_csv('http://www.math.ubc.ca/~pwalls/data/van_crime.csv')
```

```python
type(data)
```

```python
data.head()
```

The data includes homicides but the locations for these crimes are not included for privacy and the coordinates are set to 0. Let's plot the locations of all other crimes (ie. those where the coordinates are not set to 0).

```python
data = data[data['X'] > 0]
data.plot(kind='scatter',x='X',y='Y',alpha=0.1,lw=0,figsize=(12,8))
```
