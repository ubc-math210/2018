## 1. Jupyter Notebooks

### Cells

There are 2 types of cells: code cells and markdown cells. All the text we see here is contained in markdown cells. Markdown cells contain text, markdown, HTML, images, videos, whatever a browser can understand.

Code cells have Python code which executes when we hit SHIFT+ENTER and the output is displayed below.

```python
1 + 2 + 3 + 4 + 5
```

```python
0.1 + 0.2
```

### Modes

There are 2 modes: edit mode and command mode. Edit mode is when the cursor appears in the cell and we can edit the code/text in the cell. It is indicated with a green border.

Command mode is when the cursor is not in the cell and we can edit the cells in the notebook: insert, cut, copy, change cell type, etc. It is indicated by a blue border.

We can switch between modes using ESC and ENTER.

### Keyboard shortcuts

See the complete list of keyboard shortcuts in the Help menu. Here are the most common shortcuts:

| Shortcut | Command Mode Action |
| :---: | :---: |
| y | switch to code cell |
| m | switch to markdown cell |
| s | save |
| x | cut cell |
| v | paste cell below |
| c | copy cell |

## 2. Markdown

Markdown is a basically simplified HTML. The syntax allows us to include text, headings, lists, links, images and tables very easily. See [Markdown](https://daringfireball.net/projects/markdown/) for more information.

### Text

We write text as normal but we can add style:

* Use asterisks for *emphasis* (`*emphasis*`)
* Use double asterisks for **bold** (`**bold**`)
* Use back ticks for `monospace` code font (` `monospace` `)

### Headings

Use hash for headings:

* `# Heading 1`
* `## Heading 2`
* `### Heading 3`
* `#### Heading 4`
* `##### Heading 5`
* `###### Heading 6`

# Heading 1

## Heading 2

### Heading 3

#### Heading 4

##### Heading 5

###### Heading 6
